```python
def windowing(data, window_size, step_size):
    windows = []
    num_windows = (data.shape[0] - window_size) // step_size + 1
    for i in range(num_windows):
        start = i * step_size
        end = start + window_size
        windows.append(data[start:end])
    return np.array(windows)
```
